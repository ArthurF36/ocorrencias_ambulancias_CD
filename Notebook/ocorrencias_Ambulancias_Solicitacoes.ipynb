```python
import pandas as pd

df_original = pd.read_csv(r'/workspaces/ocorrencias_ambulancias_CD/Data/ocorrencias2025.csv', sep=';')

print("Dados para dataset original de ocorrências de solicitação de ambulâncias:")
display(df_original.head(10))
```

```python
print(df_original.columns.tolist())
```

```python
print(df_original.isnull().sum())
```

```python
df_dimensional = (
    df_original
    .groupby(['municipio', 'sexo'], as_index=False)
    .size()
    .rename(columns={'size': 'quantidade_solicitacoes'})
)

display(df_dimensional[df_dimensional['municipio'] == 'OLINDA'])
```

```python
df_tratado = (
    df_original
    .assign(
        municipio=lambda x: x['municipio'].fillna('Indeterminado'),
        sexo=lambda x: x['sexo'].fillna('Indeterminado')
    ))

display(df_tratado)
```

```python
df_dimensional = (
    df_tratado
    .groupby(['municipio', 'sexo'], as_index=False)
    .size()
    .rename(columns={'size': 'quantidade_solicitacoes'})
)

display(df_dimensional)
```

```python
df_dimensional.to_csv(
    "dimensional_solicitacoes_ambulancia.csv", 
    sep = ";",
    index=False,
    encoding="utf-8"
)

print("CSV dimensional criado com sucesso")
```
