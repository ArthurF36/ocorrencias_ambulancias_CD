```python
import pandas as pd

df_original = pd.read_csv(r'/workspaces/ocorrencias_ambulancias_CD/Data/ocorrencias2025.csv', sep=';')

print("=============================Extração de dados===========================")
print("Dados para dataset original de ocorrências de solicitação de ambulâncias:")
display(df_original.head(10))
```

```python
print("Visualização do nome das colunas")
print(df_original.columns.tolist())
```

```python
print("Verificação de valores nulos para cada dimensão do modelo ER (Municipio e sexo)")
print(df_original.isnull().sum())
```

```python
# Alteração dos valores nulos para "Indeterminado"
df_tratado = (
    df_original
    .assign(
        municipio=lambda x: x['municipio'].fillna('Indeterminado'),
        sexo=lambda x: x['sexo'].fillna('Indeterminado')
    ))

print("==========================Tranformação de dados===========================")
display(df_tratado)
```

```python
# Modelagem da quantidade de solicitações por dimensão
df_dimensional = (
    df_tratado
    .groupby(['municipio', 'sexo'], as_index = False)
    .size()
    .rename(columns={'size': 'quantidade_solicitacoes'})
    .sort_values(by=['quantidade_solicitacoes'], ascending = False)
)

display(df_dimensional)
```

```python
# Carregamento dos dados após transformação
df_dimensional.to_csv(
    "dimensional_solicitacoes_ambulancia.csv", 
    sep = ";",
    index=False,
    encoding="utf-8"
)

print("CSV dimensional da modelagem ER criado com sucesso")
```
